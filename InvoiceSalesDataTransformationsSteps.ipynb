Import Necessary Packages

```python
import pandas as pd
import numpy as np
import os
```

Import Excel File \
**ACTION REQUIRED**

```python
file_path = 'FILE PATH'\FY2425 A and M Inv Sales December YTD_v2.xlsx'
sheet_name = 'Invoice Details FY2324 - FY2425'
df = pd.read_excel(file_path, sheet_name=sheet_name)

df.head()
```

Retrieve All Columns

```python
for col in df.columns:
    print(col)
```

Drop Unneccessary Columns

```python
columns_to_drop = ['Effective FY',
                   'ID',
                   'Territory',
                   'Sales Person ID',
                   'Sales Person',
                   'Customer',
                   'Customer Name',
                   'GP Sold Percent',
                   'GP Sold Amount (USD)',
                   'Item Class',
                   'Catalog Unit Measure',
                   'Catalog Price',
                   'Unit Measure Sold',
                   'Price per Unit Sold',
                   'Currency',
                   'Amount Source Currency',
                   'Functional Currency',
                   'Exchange Rate To USD',
                   'Exchange Rate To Functional Currency',
                   'Tail Number',
                   'Cash/Vouch',
                   'Payment Terms',
                   'Invoice Source',
                   'Batch Number',
                   'Legal Entity',
                   'Mgt Company',
                   'Measurement QTY',
                   'OEMEnd User',
                   'Warehouse',
                   'Transactional',
                   'Russian?',
                   'Orig Order Number',
                   'Item Number',
                   'Selling Office',
                   'Invoice Number',
                   'Item Description',
                   'Item Qty'
                   ]
df = df.drop(columns=columns_to_drop)
```

Columns Used

```python
for col in df.columns:
    print(col)
```

Filtering Data | 'Type' = Actual | 'contractual/ Trans' = Contractual | 'Fin Year' = '2024-2025', '2023-2024'

```python
filtered_df = df[(df['Type'] != 'Budget') & (df['contractual/ Trans'] == 'Contractual') & (df['Fin Year'].isin(['2024-2025', '2023-2024']))]
```

Correct Filtering

```python
filtered_df.tail()
```

NA Values

```python
filtered_df.isnull().sum()
```

Ensure Correct Data Types

```python
# Get data types of all columns
column_data_types = filtered_df.dtypes

# Print the data types
print(column_data_types)
```

New CSV Data Frame Information

```python
filtered_df.info()
```

Export New CSV File for PowerBI \
**ACTION REQUIRED**

```python
# Get the directory of the original file
file_path = 'FILE PATH'/FY2425 A and M Inv Sales December YTD_v2.xlsx'
file_directory = os.path.dirname(file_path)

# Construct the output file path
output_file_path = os.path.join(file_directory, 'FY2425 A and M Inv Sales December YTD PowerBI.csv')

# Write the DataFrame to a CSV file
filtered_df.to_csv(output_file_path, index=False)

print(f"DataFrame saved to: {output_file_path}")
```
